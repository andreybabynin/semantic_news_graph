```python
 def inference(text, model, tokenizer, 
               tokenizer_kwargs = {}, 
               generate_kwargs = {}, 
               num_beams=5,):
        
    input_ids = tokenizer(
        [text],
        return_tensors="pt",
        **tokenizer_kwargs
    ).to(device)
    
    output_ids = model.generate(
        **input_ids,
        # max_length=out_max_length, #in tokens
        num_beams=num_beams,
        **generate_kwargs
    )[0]

    summary = tokenizer.decode(
        output_ids,
        skip_special_tokens=True,
    )
    return summary
```
